### check the description of the varibles 

```python
stats_desc = data_analysis.describe().transpose()
stats_desc
```

```python
stats_desc['IQR'] = stats_desc['75%'] - stats_desc['25%']
stats_desc
```

### Remove the outilers

```python
outliers = pd.DataFrame(columns=data_analysis.columns)

for col in stats_desc.index:
    iqr = stats_desc.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats_desc.at[col,'25%'] - cutoff
    upper = stats_desc.at[col,'75%'] + cutoff
    results = data_analysis[(data_analysis[col] < lower) | 
                   (data_analysis[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```
