#### Find the mean and standard deviation of the temperature of each state.

```python
np.mean(temp["Temperature"][temp["State"] == "New York"])
```

```python
# your code here
states = temp["State"].unique()

for state in states:
    mean_temperature = round(np.mean(temp['Temperature'][temp['State'] == state]), 2)
    stdev_temperature = round(np.std(temp['Temperature'][temp['State'] == state]), 2)
    print(f"Mean temperature in {state} is : {mean_temperature}")
    print(f"Standard deviation of temperature in {state} is : {stdev_temperature}")
    print("\n")
```
