#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url13 ='https://en.wikipedia.org/wiki/Python'
````

````python
# your code here
url_content13 = requests.get(url13)
soup13 = BeautifulSoup(url_content13.text, 'html.parser')

list_of_links = []
soup13.find_all('a') # Getting hold of of the blocs of links

for link in soup13.find_all('a'):
    link = link.get('href') # Getting the actual link
    list_of_links.append(link)

list_of_links
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url14 = 'http://uscode.house.gov/download/download.shtml'
````

````python
# your code here
usc_content = requests.get(url14)
soup14 = BeautifulSoup(usc_content.text, 'html.parser')

number_of_changed_titles = 0
titles_changed = soup14.find_all('div', {"class": "usctitlechanged"})

for element in titles_changed:
    number_of_changed_titles += 1
    
number_of_changed_titles
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
# your code here
content = requests.get(url)
soup = BeautifulSoup(content.text, "html.parser")
soup
````

````python
table_body = soup.find("tbody", id="tbody")
````

````python
table_head = soup.find("thead")

columns = [column_name.text for column_name in table_head.find_all("th")]

colnames = ["Date", "Time"]
````

````python
datetime = [datetime.text.strip().split() for datetime in soup.select("tbody td.tabev6 a")]
dates = [date[0] for date in datetime]
times = [time[1] for time in datetime]
````

````python
regions = [region.text.strip() for region in soup.select("tbody td.tb_region")]
````

````python
dict_earthquake = {
    'Date': dates,
    'Time': times,
    'Region': regions
}

df = pd.DataFrame(dict_earthquake)
df
````
