### 2. Use crosstabs to show the number of moovies per country

```python
movies_per_country = pd.crosstab(movies['country'], movies['title_year'], margins=True, margins_name='Total')
movies_per_country[['Total']]
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
movies_director_genre = pd.crosstab([movies['director_name'], movies["genres"]], movies['title_year'], margins=True, margins_name='Total')
movies_director_genre[['Total']]
```

### 4. Average duration of filmes grouped by color

```python
movies_color = pd.crosstab(movies['genres'], movies['color'], values=movies.duration, aggfunc='mean', margins_name='Total')
movies_color
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movies['likes']= movies['actor_1_facebook_likes'] + movies['actor_2_facebook_likes'] + movies['actor_3_facebook_likes']
```

```python
likes = pd.crosstab(movies.director_name, movies.likes, values=movies['likes'], aggfunc='sum', margins=True,  margins_name="Total" )
likes
```
