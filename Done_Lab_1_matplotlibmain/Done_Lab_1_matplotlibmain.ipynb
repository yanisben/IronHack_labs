#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
a = x ** 2
b = np.exp(x)

plt.plot(x, a, label='quadratic') # plotting a separately 
plt.plot(x, b, label='exponential') # plotting b separately 
plt.legend()
plt.show()

fig, [fig1, fig2] = plt.subplots(1,2)

fig1.plot(x, a)
fig1.set_title('LINEAR PLOT')   
fig1.set_xlabel('x label')     
fig1.set_ylabel('y label')
fig1.set_xscale("linear")

fig2.plot(x, b)
fig2.set_title('COSINE PLOT')
fig2.set_xlabel('x label')
fig2.set_ylabel('y label')
fig1.set_xscale("log")

plt.legend()
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_median = fitbit2[['Steps', 'Months_encoded']].groupby(["Months_encoded"]).agg('median')
print(steps_median)
steps_median.plot.bar()
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_week = fitbit2[['Steps', 'Months_encoded', "Work_or_Weekend"]].groupby(["Work_or_Weekend", 'Months_encoded']).agg('median')
print(steps_week)
steps_week.plot.bar()
plt.show()
```
